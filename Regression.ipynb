# linear regression

```python
#we first draw a scatter plot  
import matplotlib.pyplot as plt
x=[5,6,7,3,9,1,10,14,16]
y=[80,30,56,67,111,45,98,78,60]
plt.scatter(x,y)
plt.show()
```

```python
#draw a line of linear regression by importing scipy
import matplotlib.pyplot as plt
from scipy import stats
x=[5,7,8,7,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,111,86,103,87,94,78,77,85,86]
#execute the method that returns important key of linear regression
slope, intercept,r,p,std_err=stats.linregress(x,y)
#create a function that uses slope and intercept values to return new values
#y=mx+c
def myfunc(x):
    return slope*x +intercept
#run each value of the x array through the function created above
mymodel=list(map(myfunc,x))
plt.scatter(x,y)
#draw the line of linear regression
plt.plot(x,mymodel)
plt.show()

```

# polynomial regression

```python
#draw the scatter plot
import matplotlib.pyplot as plt
x=[1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y=[100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
plt.scatter(x,y)
plt.show()
```

```python
#draw the line of polynomial regression
import matplotlib.pyplot as plt
import numpy 
x=[1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y=[100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
#it is a method that lets us make a polynomial model
model=numpy.poly1d(numpy.polyfit(x,y,3))
#specify how the line will display ,where we start from 1 and end to 22
line=numpy.linspace(1,22,100)
plt.scatter(x,y)
#draw the line of polynomial regression
plt.plot(line,model(line))
plt.show()
```

```python
#how well the data fit in regression
import numpy
from sklearn.metrics import r2_score
x=[1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y=[100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
model=numpy.poly1d(numpy.polyfit(x,y,3))
print(r2_score(y,model(x)))
```
